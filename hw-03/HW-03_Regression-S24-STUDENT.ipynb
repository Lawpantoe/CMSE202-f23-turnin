# Part 3: One Variable Linear Regression (30 points)

In this part, we'll perform some one-variable linear regression analysis on the world happiness data.

&#9989;  **Question 3.1 (6 points)**: Using the `OLS` method in `statsmodels`, perform a linear regression to predict the `happiness_score` using the `healthy_life_expectancy` column separately for 2022 and 2023 and **display the results summaries** for both years. Remember that you may need to use the `add_constant()` method to make sure `OLS` fits a general line $y = ax+b$ to the data instead of a line through the origin $y = ax$. 

````python
# Put your code here
import statsmodels.api as sm
import matplotlib.pyplot as plt
#2022
y= WHR_2022["happiness_score"]
x= WHR_2022["healthy_life_expectancy"]

#OLS regression 2022
x_with_cnst = sm.add_constant(x)
model = sm.OLS(y, x_with_cnst)
results = model.fit()
print("Intercept and slope are:", results.params)
print(results.summary())

print()

#2023
y1= WHR_2023["happiness_score"]
x1= WHR_2023["healthy_life_expectancy"]

#OLS regression 2023
x_with_cnst1 = sm.add_constant(x1)
model1 = sm.OLS(y1, x_with_cnst1)
results1 = model1.fit()
print("Intercept and slope are:", results1.params)
print(results1.summary())
````

&#9989;  **Question 3.3 (6 points)**: Now make scatter plots of `happiness_score` (x-axis) vs. `healthy_life_expectancy` (on y-axis) for 2022 and 2023 on the same plot.
````python
# Put your code here

# Extracting coefficients for 2022
intercept_2022 = results.params.iloc[0]
slope_2022 = results.params.iloc[1]

# Scatter plot and best-fit line for 2022
plt.scatter(WHR_2022["happiness_score"], WHR_2022["healthy_life_expectancy"], label="2022",color='blue')
plt.plot(WHR_2022["happiness_score"], slope_2022 * WHR_2022["happiness_score"] + intercept_2022, color='blue')

# Extracting coefficients for 2023
intercept_2023 = results1.params.iloc[0]
slope_2023 = results1.params.iloc[1]

# Scatter plot and best-fit line for 2023
plt.scatter(WHR_2023["happiness_score"], WHR_2023["healthy_life_expectancy"], label="2023",color='orange')
plt.plot(WHR_2023["happiness_score"], slope_2023 * WHR_2023["happiness_score"] + intercept_2023 ,color='orange')

plt.xlabel("Happiness Score")
plt.ylabel("Healthy Life Expectancy")
plt.title("Happiness Score vs Healthy Life Expectancy")
plt.legend()
plt.show()
````
